```python
import pandas as pd
import numpy as np
np.random.seed(42)

# 生成2019-2023年的数据
years = range(2019, 2024)
brands = ['可口可乐', '百事可乐', '农夫山泉', '康师傅', '统一']

data = []
for brand in brands:
    # 为每个品牌设置基础销售额和增长趋势
    base = np.random.uniform(8000, 15000)
    growth = np.random.uniform(1.03, 1.08)
    
    for year in years:
        # 添加随机波动
        sales = base * (growth ** (year - 2019)) * np.random.uniform(0.95, 1.05)
        data.append({
            '品牌': brand,
            '年份': year,
            '销售额': round(sales, 2)  # 单位：百万元
        })

# 创建DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate data
years = range(2019, 2024)
brands = ['Coca Cola', 'Pepsi', 'Nongfu Spring', 'Master Kong', 'Uni-President']

data = []
for brand in brands:
   base = np.random.uniform(8000, 15000)
   growth = np.random.uniform(1.03, 1.08)
   
   for year in years:
       sales = base * (growth ** (year - 2019)) * np.random.uniform(0.95, 1.05)
       data.append({
           'Brand': brand,
           'Year': year,
           'Sales': round(sales, 2)
       })

df = pd.DataFrame(data)

# Line plot: Sales trend
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Sales', hue='Brand', marker='o')
plt.title('Sales Trend by Brand')
plt.show()

# Stacked bar plot: Brand composition
plt.figure(figsize=(12, 6))
sales_by_year = df.pivot(index='Year', columns='Brand', values='Sales')
sales_by_year.plot(kind='bar', stacked=True)
plt.title('Annual Sales Composition by Brand')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()

# Heatmap: Sales intensity
plt.figure(figsize=(10, 6))
pivot_table = df.pivot(index='Brand', columns='Year', values='Sales')
sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='YlOrRd')
plt.title('Sales Heatmap')
plt.show()

# Pie chart: Market share
plt.figure(figsize=(8, 8))
latest_year = df[df['Year'] == 2023]
plt.pie(latest_year['Sales'], labels=latest_year['Brand'], autopct='%1.1f%%')
plt.title('Market Share 2023')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate data 
years = range(2019, 2024)
brands = ['Coca Cola', 'Pepsi', 'Nongfu Spring', 'Master Kong', 'Uni-President']
data = []

for brand in brands:
   base = np.random.uniform(8000, 15000)
   growth = np.random.uniform(1.03, 1.08)
   for year in years:
       sales = base * (growth ** (year - 2019)) * np.random.uniform(0.95, 1.05)
       data.append({
           'Brand': brand,
           'Year': year,
           'Sales': round(sales, 2)
       })

df = pd.DataFrame(data)

# Create visualization
fig, ax = plt.subplots(figsize=(15, 8), facecolor='white')
ax.set_facecolor('white')

for i, brand in enumerate(brands):
   brand_data = df[df['Brand'] == brand]
   # Plot scatter points
   ax.scatter(brand_data['Year'], brand_data['Sales'], 
             s=100, alpha=0.6, label=brand)
   # Plot connecting lines
   ax.plot(brand_data['Year'], brand_data['Sales'], 
          alpha=0.4, linewidth=0.8)
   
   # Add vertical lines
   for _, row in brand_data.iterrows():
       ax.vlines(x=row['Year'], ymin=0, ymax=row['Sales'], 
                alpha=0.1, linewidth=0.5)

ax.set_title('Brand Sales Trends (2019-2023)', pad=20, fontsize=14)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate data
years = range(2019, 2024)
brands = ['Coca Cola', 'Pepsi', 'Nongfu Spring', 'Master Kong', 'Uni-President']
data = []

for brand in brands:
   base = np.random.uniform(8000, 15000)
   growth = np.random.uniform(1.03, 1.08)
   for year in years:
       sales = base * (growth ** (year - 2019)) * np.random.uniform(0.95, 1.05)
       data.append({
           'Brand': brand,
           'Year': year,
           'Sales': round(sales, 2)
       })

df = pd.DataFrame(data)

# Create 3D visualization
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Convert brands to numeric values for plotting
brand_nums = pd.Categorical(df['Brand']).codes

# Plot 3D scatter
scatter = ax.scatter(df['Year'], 
                   brand_nums,
                   df['Sales'],
                   c=df['Sales'],
                   cmap='viridis',
                   s=100)

# Customize the plot
ax.set_yticks(range(len(brands)))
ax.set_yticklabels(brands)
ax.set_xlabel('Year')
ax.set_ylabel('Brand')
ax.set_zlabel('Sales')
ax.view_init(elev=20, azim=45)

# Add colorbar
plt.colorbar(scatter)
plt.show()
```

```python
import pandas as pd
import numpy as np
from ipywidgets import interactive
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 生成数据
years = range(2019, 2024)
brands = ['Coca Cola', 'Pepsi', 'Nongfu Spring', 'Master Kong', 'Uni-President']
data = []

for brand in brands:
    base = np.random.uniform(8000, 15000)
    growth = np.random.uniform(1.03, 1.08)
    for year in years:
        sales = base * (growth ** (year - 2019)) * np.random.uniform(0.95, 1.05)
        data.append({
            'Brand': brand,
            'Year': year,
            'Sales': round(sales, 2)
        })

df = pd.DataFrame(data)

def plot_3d(elevation, azimuth):
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    for i, brand in enumerate(brands):
        brand_data = df[df['Brand'] == brand]
        ax.scatter(brand_data['Year'], 
                  [i] * len(brand_data), 
                  brand_data['Sales'],
                  label=brand)
        
    ax.set_yticks(range(len(brands)))
    ax.set_yticklabels(brands)
    ax.set_xlabel('Year')
    ax.set_ylabel('Brand')
    ax.set_zlabel('Sales')
    
    ax.view_init(elev=elevation, azim=azimuth)
    plt.legend()
    plt.show()

interactive_plot = interactive(plot_3d, 
                             elevation=(-90, 90, 5),
                             azimuth=(-180, 180, 5))
display(interactive_plot)
```
